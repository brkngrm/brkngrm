```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
html = requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250').content
print(html)
```

```python
soup = BeautifulSoup(html,'html.parser')
print(soup.tbody)
```

```python
rows = soup.tbody.find_all('tr')
print(rows[:1])
for row in rows:
    column = row.find('td', 'titleColumn')
    print(column.text)
```

```python
list1 = []
for row in rows:
    column = row.find('td',"titleColumn")
    title = column.a.text
    year = column.span.text
    column1 = row.find('td', 'ratingColumn imdbRating')
    rating = column1.strong.text
    string = (f'{title} - {year} - {rating}')
    list1.append(string)
def extractDigits(list1):
   return [[movie] for movie in list1]
final = extractDigits(list1)
```

```python
df = pd.DataFrame(final)
df.head()
```

```python
df1 = df[0].str.split('-', expand=True)
df1.pop(4)
df1.pop(3)
df1.rename(columns={0:'Movie Title',1:'Year Released',2:'IMDb Rating'}, inplace=True)
df1.head(10)
```
